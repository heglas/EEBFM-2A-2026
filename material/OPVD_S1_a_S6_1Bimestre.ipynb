# Semana 4 - Funcoes Condicionais e Logicas
**ANO1C3B1S4 - Aulas 1, 2 e 3**

Permitem que o Excel **tome decisoes automaticas** baseadas em testes logicos.

## Funcao SE e Operadores Relacionais (S4A1)

| Operador | Significado |
|---|---|
| = | Igual a |
| <> | Diferente de |
| > | Maior que |
| < | Menor que |
| >= | Maior ou igual |
| <= | Menor ou igual |

**Estrutura:** `=SE(teste_logico; verdadeiro; falso)`

**Exemplo:** `=SE(B2>=7; "Aprovado"; "Reprovado")`

## SE com E, OU, NAO (S4A2)

| Operador | Retorna VERDADEIRO quando... |
|---|---|
| E(c1;c2) | TODAS as condicoes sao verdadeiras |
| OU(c1;c2) | PELO MENOS UMA condicao e verdadeira |
| NAO(c) | A condicao e FALSA (inverte) |

`=SE(E(B2>=7; C2>=75%); "Aprovado"; "Reprovado")`


### Simulacao - SE, E, OU

````
import pandas as pd

alunos = pd.DataFrame({
    'Nome':       ['Ana','Bruno','Carla','Daniel','Elisa','Felipe'],
    'Nota':       [8.0, 5.5, 9.0, 6.5, 4.0, 7.0],
    'Frequencia': [0.85, 0.70, 0.90, 0.80, 0.65, 0.75]
})
alunos['SE_simples'] = alunos['Nota'].apply(
    lambda n: 'Aprovado' if n >= 7 else 'Reprovado')
alunos['SE_E'] = alunos.apply(
    lambda r: 'Aprovado' if (r['Nota']>=7 and r['Frequencia']>=0.75) else 'Reprovado', axis=1)
alunos['SE_OU'] = alunos.apply(
    lambda r: 'Destaque' if (r['Nota']>=9 or r['Frequencia']>=0.90) else 'Normal', axis=1)
print(alunos.to_string(index=False))
````

### Simulacao - Relatorio Escolar Automatizado (S4A3)

````
import pandas as pd

alunos2 = pd.DataFrame({
    'Nome':     ['Ana','Bruno','Carla','Daniel','Elisa','Felipe','Gabi'],
    'Nota':     [8.0, 5.5, 9.0, 4.0, 6.5, 7.0, 3.0],
    'Freq_pct': [85, 70, 90, 80, 65, 75, 55]
})

def classificar(row):
    n, f = row['Nota'], row['Freq_pct']
    if n < 5 or f < 60:      return 'Reprovado'
    elif n >= 7 and f >= 75: return 'Aprovado'
    else:                    return 'Recuperacao'

alunos2['Situacao'] = alunos2.apply(classificar, axis=1)
print(alunos2.to_string(index=False))
print()
print(alunos2['Situacao'].value_counts().to_string())
````

# Semana 6 - Funcoes de Busca e Referencia
**ANO1C3B1S6 - Aulas 4, 5 e 6**

Cruzar informacoes entre tabelas de forma automatica.

## PROCV e PROCH (S6A1)

**PROCV — Busca Vertical**
```
=PROCV(valor_procurado; tabela; numero_coluna; FALSO)
```

**PROCH — Busca Horizontal**
```
=PROCH(valor_procurado; tabela; numero_linha; FALSO)
```

> PROCV: tabela cresce para baixo | PROCH: tabela cresce para a direita

## INDICE e CORRESP (S6A2)

**Problema do PROCV:** so funciona se a coluna de busca for a **primeira** da tabela.

```
=CORRESP(valor; intervalo; 0)      → retorna a POSICAO
=INDICE(coluna; posicao)           → retorna o VALOR

Combinacao:
=INDICE(col_resultado; CORRESP(valor; col_busca; 0))
```

Funciona **independente da ordem das colunas**!

## Comparativo Final (S6A3)

| Criterio | PROCV | PROCH | INDICE+CORRESP |
|---|---|---|---|
| Direcao | Vertical | Horizontal | Qualquer |
| Coluna de busca | Deve ser a 1a | Deve ser a 1a linha | Qualquer |
| Flexibilidade | Baixa | Baixa | Alta |
| Quando reorganiza a tabela | Quebra | Quebra | Continua funcionando |


### Simulacao - PROCV

````
import pandas as pd

produtos = pd.DataFrame({
    'CodProd':   ['P001','P002','P003','P004','P005'],
    'Produto':   ['Notebook','Mouse USB','Teclado','Monitor 24pol','Webcam HD'],
    'Categoria': ['Eletronico','Periferico','Periferico','Eletronico','Periferico'],
    'Preco':     [3500.00, 89.90, 199.90, 1299.00, 299.90]
})

def procv(v, df, col_busca, col_ret):
    r = df[df[col_busca]==v][col_ret]
    return r.values[0] if len(r)>0 else '#N/D'

print(produtos.to_string(index=False))
print()
print(f"{'Codigo':<8} {'Produto':<18} {'Preco'}")
print("-"*40)
for cod in ['P003','P001','P999','P004']:
    print(f"{cod:<8} {str(procv(cod,produtos,'CodProd','Produto')):<18} {str(procv(cod,produtos,'CodProd','Preco'))}")
````

### Simulacao - INDICE + CORRESP

````
import pandas as pd

# Codigo NAO e a 1a coluna - PROCV quebraria!
clientes = pd.DataFrame({
    'Valor_Total': [1500.00, 3200.50, 870.00, 4500.00, 220.00],
    'Nome':        ['Ana Lima','Bruno Costa','Carla Dias','Delta Corp','Eder Silva'],
    'Codigo':      ['C001','C002','C003','C004','C005'],
    'Cidade':      ['SP','RJ','MG','SP','PR']
})

def ind_corresp(v, col_busca, col_ret, df):
    try:
        pos = list(df[col_busca]).index(v)
        return df[col_ret].iloc[pos]
    except:
        return '#N/D'

print("ATENCAO: Codigo e a 3a coluna - PROCV nao funcionaria!")
print(clientes.to_string(index=False))
print()
print(f"{'Codigo':<8} {'Nome':<15} {'Valor Total'}")
print("-"*38)
for cod in ['C003','C001','C004','C999']:
    n = ind_corresp(cod,'Codigo','Nome',clientes)
    v = ind_corresp(cod,'Codigo','Valor_Total',clientes)
    print(f"{cod:<8} {str(n):<15} {str(v)}")
````

````
# Remover duplicatas e padronizar nomes
cadastro_limpo = cadastro.drop_duplicates(subset='RA').copy()
cadastro_limpo['Nome'] = cadastro_limpo['Nome'].str.title()

erros_nota = cadastro_limpo[~cadastro_limpo['Nota_B1'].between(0, 10)]
erros_freq = cadastro_limpo[~cadastro_limpo['Freq_pct'].between(0, 100)]

print("=== CADASTRO LIMPO ===")
print(cadastro_limpo.to_string(index=False))
print()
print("=== DADOS INVALIDOS ENCONTRADOS ===")
if len(erros_nota) > 0:
    print("Notas invalidas:")
    print(erros_nota[['RA','Nome','Nota_B1']].to_string(index=False))
if len(erros_freq) > 0:
    print("Frequencias invalidas:")
    print(erros_freq[['RA','Nome','Freq_pct']].to_string(index=False))
````

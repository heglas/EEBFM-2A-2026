### 2.1 — Lógica Booleana
Em Python, valores `True` e `False` são do tipo `bool`. Operadores lógicos combinam expressões booleanas.

### Tabela-Verdade

| A | B | A and B | A or B | not A |
|---|---|---------|--------|-------|
| True | True | **True** | **True** | False |
| True | False | **False** | **True** | False |
| False | True | **False** | **True** | **True** |
| False | False | **False** | **False** | **True** |

**Regras de ouro:**
- `and` → só é `True` se **ambos** forem verdadeiros
- `or` → é `True` se **pelo menos um** for verdadeiro
- `not` → **inverte** o valor booleano


````python
# Tabela-verdade na prática
print("=== Operador AND ===")
print(True and True)   # True
print(True and False)  # False
print(False and True)  # False
print(False and False) # False

print("\n=== Operador OR ===")
print(True or True)    # True
print(True or False)   # True
print(False or True)   # True
print(False or False)  # False

print("\n=== Operador NOT ===")
print(not True)        # False
print(not False)       # True
````

### 2.2 — Aplicações Reais com `and`, `or`, `not`

Contexto: Sistema de matrícula da EEBFM — verificar se um aluno pode ser aprovado.


````python
# Aprovação por nota E frequência (AND)
nota = 7.0
frequencia = 80  # percentual

aprovado_nota = nota >= 6
aprovado_freq  = frequencia >= 75
aprovado_final = aprovado_nota and aprovado_freq

print(f"Nota ({nota}) >= 6: {aprovado_nota}")
print(f"Frequência ({frequencia}%) >= 75%: {aprovado_freq}")
print(f"Aprovado no bimestre: {aprovado_final}")
````

````python
# Acesso ao laboratório (OR)
e_professor = False
tem_autorizacao = True
pode_entrar = e_professor or tem_autorizacao
print("Pode entrar no laboratório:", pode_entrar)

# Verificar se é menor de idade (NOT)
idade = 17
eh_maior = idade >= 18
eh_menor  = not eh_maior
print(f"\nIdade {idade} → É maior de idade: {eh_maior} | É menor: {eh_menor}")
````

### 2.3 — Expressões Booleanas Complexas

Python avalia operadores na ordem: `not` → `and` → `or`  
Use parênteses para tornar a lógica mais clara.


````python
# Expressão complexa: frete grátis
valor_compra = 150
tem_cupom    = False
eh_membro    = True

# Frete grátis se: compra >= 200 OU (cupom E membro)
frete_gratis = (valor_compra >= 200) or (tem_cupom and eh_membro)
print("Frete grátis:", frete_gratis)

# Acesso a jogo online
idade = 16
tempo_de_jogo = 2  # horas hoje
saldo = 30.0

pode_jogar = (idade >= 12) and (tempo_de_jogo < 4) and (saldo > 0)
print("Pode jogar:", pode_jogar)
````

## 🔀 Bloco 3 — Estruturas Condicionais: `if`, `if-else`, `if-elif-else`
### Semana 6 | Código: DADOSANO1C2B1S6


### 3.1 — `if` (Condicional Simples)

```python
if <condição>:
    # bloco executado somente se a condição for True
```

⚠️ **Atenção:** A indentação (4 espaços) define o bloco. Sem ela, erro de sintaxe!


````python
# if simples
temperatura = 38.5

if temperatura > 37.5:
    print("⚠️  Temperatura acima do normal! Procure um médico.")

print("Verificação concluída.")
````

### 3.2 — `if-else` (Condicional Composta)

```python
if <condição>:
    # executa se True
else:
    # executa se False
```

Com `if-else` **sempre** haverá uma resposta, independente do valor.


````python
# if-else
media = 5.5

if media >= 6:
    print("✅ Aprovado!")
else:
    print("❌ Reprovado — vá para recuperação.")
````

````python
# if-else com saldo bancário
saldo = 250.00
valor_saque = 300.00

if saldo >= valor_saque:
    saldo -= valor_saque
    print(f"Saque de R${valor_saque:.2f} realizado. Saldo atual: R${saldo:.2f}")
else:
    print(f"Saldo insuficiente! Saldo: R${saldo:.2f}, Tentativa: R${valor_saque:.2f}")
````

### 3.3 — `if-elif-else` (Múltiplas Condições)

Quando temos **mais de dois** resultados possíveis, usamos `elif` (abreviação de *else if*).

```python
if <condição 1>:
    ...
elif <condição 2>:
    ...
elif <condição 3>:
    ...
else:
    ...
```

⚠️ Python testa de cima para baixo e **para no primeiro `True`** encontrado.


````python
# Classificação de média — EEBFM
media = 7.2

if media >= 8.5:
    conceito = "Excelente"
elif media >= 7.0:
    conceito = "Bom"
elif media >= 6.0:
    conceito = "Regular — Aprovado"
elif media >= 4.0:
    conceito = "Recuperação"
else:
    conceito = "Reprovado"

print(f"Média: {media} → Conceito: {conceito}")
````

````python
# Classificação de IMC
peso = 70  # kg
altura = 1.75  # m

imc = peso / (altura ** 2)
print(f"IMC calculado: {imc:.2f}")

if imc < 18.5:
    classificacao = "Abaixo do peso"
elif imc < 25.0:
    classificacao = "Peso normal"
elif imc < 30.0:
    classificacao = "Sobrepeso"
else:
    classificacao = "Obesidade"

print(f"Classificação: {classificacao}")
````

### 3.4 — Condicionais Aninhadas

Um `if` dentro de outro `if`. Use com moderação — em excesso dificulta a leitura.


````python
# Controle de acesso por faixa etária e horário
idade = 16
horario = 20  # horas

if idade >= 18:
    print("Acesso permitido para adulto.")
else:
    if horario < 18:
        print("Acesso permitido para menor durante o dia.")
    else:
        print("⚠️  Acesso negado: menores não podem entrar após as 18h sem responsável.")
````

### Desafio 4.1 — Sistema de Matrícula
Crie um sistema que verifica se um aluno pode se matricular no Técnico em Ciência de Dados:
- Deve ter idade ≥ 14 anos  
- Deve ter concluído o ensino fundamental  
- Deve ter nota de redação ≥ 5.0 OU média escolar ≥ 7.0


````python
# Desafio 4.1 — Sistema de Matrícula EEBFM
print("=== Sistema de Matrícula — Técnico em Ciência de Dados ===")
print("E.E. Profa. Benedita Freire de Macedo\n")

nome = input("Nome do candidato: ")
idade = int(input("Idade: "))
concluiu_ef = input("Concluiu o Ensino Fundamental? (s/n): ").lower() == "s"
nota_redacao = float(input("Nota de Redação (0-10): "))
media_escolar = float(input("Média Escolar (0-10): "))

# Lógica de elegibilidade
criterio_idade  = idade >= 14
criterio_ef     = concluiu_ef
criterio_notas  = (nota_redacao >= 5.0) or (media_escolar >= 7.0)

pode_matricular = criterio_idade and criterio_ef and criterio_notas

print(f"\n--- Resultado para {nome} ---")
print(f"Critério de idade (>= 14): {criterio_idade}")
print(f"Ensino Fundamental concluído: {criterio_ef}")
print(f"Critério de notas (redação>=5 OU média>=7): {criterio_notas}")

if pode_matricular:
    print("\n✅ MATRICULADO! Bem-vindo ao Técnico em Ciência de Dados!")
else:
    print("\n❌ Não elegível para matrícula no momento.")
    if not criterio_idade:
        print("   → Motivo: Idade mínima não atingida.")
    if not criterio_ef:
        print("   → Motivo: Ensino Fundamental não concluído.")
    if not criterio_notas:
        print("   → Motivo: Critério de notas não atendido.")
````

### Desafio 4.2 — Calculadora de Conceitos por Bimestre
Simule o sistema de avaliação completo da EEBFM com todos os bimestres.


````python
# Desafio 4.2 — Calculadora completa
print("=== Calculadora de Aprovação Anual — EEBFM ===\n")

notas = []
for i in range(1, 5):
    n = float(input(f"Nota do {i}º Bimestre: "))
    notas.append(n)

media_anual = sum(notas) / len(notas)
frequencia  = float(input("Frequência anual (%): "))

print(f"\nMédia Anual: {media_anual:.2f}")
print(f"Frequência: {frequencia:.1f}%")

if frequencia < 75:
    resultado = "Reprovado por Falta"
elif media_anual >= 6.0:
    resultado = "Aprovado"
elif media_anual >= 4.0:
    nota_rec = float(input("\nEm recuperação! Digite a nota da recuperação: "))
    media_rec = (media_anual + nota_rec) / 2
    print(f"Média após recuperação: {media_rec:.2f}")
    resultado = "Aprovado após Recuperação" if media_rec >= 5.0 else "Reprovado"
else:
    resultado = "Reprovado"

print(f"\n📋 Resultado Final: {resultado}")
````

## ⚠️ Bloco 6 — Erros Comuns e Como Evitar


### Erros mais frequentes observados em sala

| # | Erro | Exemplo Errado | Correto |
|---|------|----------------|---------|
| 1 | Usar `=` em vez de `==` | `if nota = 6:` | `if nota == 6:` |
| 2 | Esquecer os dois-pontos `:` | `if nota >= 6` | `if nota >= 6:` |
| 3 | Indentação errada | bloco fora do `if` | 4 espaços após `if` |
| 4 | Ordem errada no `elif` | verificar `>=6` antes de `>=8` | verificar do maior para o menor |
| 5 | `not` com parênteses esquecidos | `not x > 5` ≠ `not(x > 5)` | usar parênteses explícitos |


````python
# Demonstração do Erro 4 — Ordem do elif (ERRADO)
media = 9.0

# ERRADO: nunca chegará ao elif de excelente
if media >= 6.0:
    print("ERRADO → Aprovado")  # Este sempre "ganha"
elif media >= 8.5:
    print("Excelente")  # Nunca executado!
````

````python
# CORRETO: do maior para o menor
media = 9.0

if media >= 8.5:
    print("✅ Excelente")
elif media >= 6.0:
    print("Aprovado")
else:
    print("Reprovado")
````
